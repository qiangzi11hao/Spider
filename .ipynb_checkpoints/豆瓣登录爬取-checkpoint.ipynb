爬取IMDB TOP 250

```python
s = requests.Session()
contents=[]
a=""
for id in range(0,251,25):
    url = 'https://www.douban.com/doulist/1518184/?start='+str(id)
    r = s.get(url)
    r.encoding = 'utf-8'
    root = BeautifulSoup(r.text)
    #print(root)
    #items = root.select('#item8900689')
    items = root.select('#content > div > div.article > div.doulist-item')
    #print(len(items))
    
    for item in items:
        try:
            title = item.select(' div > div.bd.doulist-subject > div.title > a')[0].string
            rank = item.select(' div > div.bd.doulist-subject > div.rating > span.rating_nums')[0].string
            a += str(title.strip()) +' :' +str(rank) +'\n'
#             contents.append(str(title.strip()))
#             contents.append(':')
#             contents.append(str(rank))
#             contents.append('\n')
        except:
            continue
        
# for content in contents:
#     a+=content
print(a)
with open('douban_movie_top250.txt','wb+') as f:
    f.write(bytes(a.encode()))
f.close()
```
